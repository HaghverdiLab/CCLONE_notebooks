```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
import seaborn as sns
import anndata

from itertools import permutations
import sklearn.metrics


plt.rcParams['pdf.fonttype'] = 42 #for saving PDF with changeable text
plt.rcParams['ps.fonttype'] = 42 #for saving PDF with changeable text

# my code
%load_ext autoreload
%autoreload 2
from passenger.plot.plot import *


A_patients = ["A"+str(i) for i in np.arange(1,16)]
P_patients = ["P1", "P2", "P3", "P4"]
LUNG_patients = ["p018", "p019", "p023", "p030", "p031", "p032", "p033"]

all_patients = np.concatenate([P_patients, A_patients, LUNG_patients])
```

```python
mean_A_patients = []
for A in A_patients:
    data = pd.read_csv("../../../raw_data/"+A+"/reads_per_barcode.txt", sep="\s", header=None, index_col=1)
    adata = anndata.read_h5ad("../processed_data/AML_10X/"+A+".h5ad")
    cells = adata.obs.index.str.split("-").str[0]
    data = data.loc[cells]
    mean_A_patients.append(np.mean(data[0]))
    print(mean_A_patients)
```

```python
mean_LUNG_patients = []
for A in LUNG_patients:
    data = pd.read_csv("../../../raw_data/"+A+"/reads_per_barcode.txt", sep="\s", header=None, index_col=1)
    adata = anndata.read_h5ad("../processed_data/LUNG/"+A+".h5ad")
    cells = adata.obs.index.str.split("-").str[0]
    data = data.loc[cells]
    mean_LUNG_patients.append(np.mean(data[0]))
    print(A, np.mean(data[0]))
```

```python
fig, ax = plt.subplots(1,1, figsize=(10, 10))


df = pd.read_csv("../../../raw_data/n_reads_per_patient.tsv", sep="\t", index_col=0)

sub_df = df.loc[df.index.str[0]=="P"]
plt.scatter((sub_df["n_reads"]/sub_df["n_cells"]).tolist(), sub_df["n_vars"].tolist(), marker="o", s=50, c="pink")
for i, lab in enumerate(sub_df.index):
    ax.annotate(lab, (sub_df["n_reads"].iloc[i]/sub_df["n_cells"].iloc[i], sub_df.n_vars.iloc[i]))

sub_df = df.loc[A_patients]
plt.scatter(mean_A_patients, sub_df["n_vars"].tolist(), marker="^", s=50, c="green")
for i, lab in enumerate(sub_df.index):
    ax.annotate(lab, (mean_A_patients[i], sub_df.n_vars.iloc[i]))  
    
sub_df = df.loc[LUNG_patients]
plt.scatter(mean_LUNG_patients, sub_df["n_vars"].tolist(), marker="*", s=50, c="gold")
for i, lab in enumerate(LUNG_patients):
    ax.annotate(lab, (mean_LUNG_patients[i], sub_df.n_vars.iloc[i])) 
ax.set_xticks([0, 200000, 400000, 600000, 800000, 100000])#, ["0", "2e5", "4e5", "6e5", "8e5"])
ax.set_yticks([0, 10000, 20000, 30000])#, ["0", "1e4", "2e4", "3e4"])

plt.xlim(10000,1000000)
#plt.ylim(0)
plt.xscale("log")
ax.set_xticks([10000, 100000, 1000000])

ax.get_xaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
plt.savefig("../figures/QC_read-per-cell-vs-vars.pdf", bbox_inches='tight', format='pdf', dpi=300)
plt.show()
```

```python
subsets = np.array([1, 0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1, 0.05])

precision, recall = [], []
n_vars = []
sub_patients = []
for patient in all_patients:
    print(patient)
    prefix = "AML_MutaSeq" if patient[0]=="P" else "AML_10X" if patient[0]=="A" else "LUNG"
    adata = anndata.read_h5ad("../processed_data/"+prefix+"/"+patient+".h5ad")
    
    if "cell_labels" not in adata.obs.columns:
        continue
           
    sub_cells = adata.obs_names[adata.obs["cell_labels"]!="undetermined"]
    adata = adata[sub_cells]
    cancer_factors = np.where(adata.uns['factor_labels']=="cancer")[0]
    ref_labels = [i in cancer_factors for i in  np.argmax(adata.obsm["C"], axis=1)]
    
    prec, rec, n_v = [], [], []
    try:
        for i in subsets:
            p, r,nv = [], [], []


            for j in range(5):
                subset_data = anndata.read_h5ad("../processed_data/subsampled//"+patient+"_"+str(i.astype(int) if i == 1 else i)+"_"+str(j+1)+".h5ad")[sub_cells]

                C = subset_data.obsm["C"]
                perm = list(permutations(np.arange(0, C.shape[1])))
                dists = [np.sum(np.abs((adata.obsm["C"]-C[:,perm[i]]))) for i in range(len(perm))]
                aligned = C[:,perm[np.argmin(dists)]]
                new_labels = [i in cancer_factors for i in np.argmax(aligned, axis=1)]
                p.append(sklearn.metrics.precision_score(ref_labels, new_labels))
                r.append(sklearn.metrics.recall_score(ref_labels, new_labels))
                p.append(sklearn.metrics.precision_score(ref_labels, new_labels))
                r.append(sklearn.metrics.recall_score(ref_labels, new_labels))
                nv.append(subset_data.shape[1])
            prec.append(p)
            rec.append(r)
            n_v.append(nv)
        precision.append(prec)
        recall.append(rec)
        n_vars.append(n_v)
    except:
        print(patient, i, j)
        continue
              
    sub_patients.append(patient)
precision=np.array(precision)
recall=np.array(recall) 
n_vars = np.array(n_vars)
```

```python
mean_p = np.mean(precision, axis=2)
mean_r = np.mean(recall, axis=2)

std_p = np.std(precision, axis=2)
std_r = np.std(recall, axis=2)

mean_n_vars = np.mean(n_vars, axis=2)
std_n_vars = np.std(n_vars, axis=2)
```

```python
fig, ax = plt.subplots(1,1,figsize=(5,5))

x = (1-subsets)

for i, patient in enumerate(sub_patients):
    c = np.repeat("lightgrey", 11)
    colors_sub = mean_n_vars[i]>100
    c[colors_sub]="pink" if patient[0]=="P" else "green" if patient[0]=="A" else "gold"

        
    y, e = mean_p[i], std_p[i]
    
    ax.annotate(patient, (x[colors_sub][-1], y[colors_sub][-1]))  
    ax.scatter(x, y, c=c.tolist())
    ax.errorbar(x[colors_sub],y[colors_sub], e[colors_sub],c=c[0], alpha=.5)
    if np.any(~colors_sub):
        loc = np.where(~colors_sub)[0][0]-1
        ax.plot(x[loc:],y[loc:], c="lightgrey")
    
plt.ylim(0, 1)
plt.xticks(1-subsets, [str(i) for i in np.round((subsets)*100).astype(int)])
plt.savefig("../figures/QC_prec.pdf", bbox_inches='tight', format='pdf', dpi=300)
plt.show()

fig, axs = plt.subplots(1,2,figsize=(10,5))

for i, patient in enumerate(sub_patients):
    c = np.repeat("lightgrey", 11)
    colors_sub = mean_n_vars[i]>100
    c[colors_sub]="pink" if patient[0]=="P" else "green" if patient[0]=="A" else "gold"
            
    y, e = mean_r[i], std_r[i]
    axs[0].annotate(patient, (x[colors_sub][-1], y[colors_sub][-1]))  
    axs[0].scatter(x, y, c=c.tolist())
    axs[0].errorbar(x[colors_sub],y[colors_sub], e[colors_sub],c=c[0], alpha=.5)
    if np.any(~colors_sub):
        loc = np.where(~colors_sub)[0][0]-1
        axs[0].plot(x[loc:],y[loc:], c="lightgrey")
    axs[0].set_ylabel("Recall")
    axs[0].set_ylim(0,1)
    
    y, e = mean_n_vars[i], std_n_vars[i]
    axs[1].annotate(patient, (x[0]-0.05, y[0]))  
    axs[1].scatter(x, y, c=c.tolist())
    axs[1].plot(x[colors_sub],y[colors_sub], c=c[0], alpha=.5)
    axs[1].errorbar(x[colors_sub],y[colors_sub], e[colors_sub],c=c[0], alpha=.5)
    if np.any(~colors_sub):
        axs[1].plot(x[np.where(~colors_sub)[0][0]-1:],y[np.where(~colors_sub)[0][0]-1:], c=c[-1], alpha=.5)
    axs[1].set_ylabel("Number of variants")
    
for ax in axs:
    ax.set_xlabel("Percentage sampled counts")
    ax.set_xticks(1-subsets, [str(i) for i in np.round((subsets)*100).astype(int)])
    ax.set_xlim(-0.1, 1)
    
plt.xticks(1-subsets, [str(i) for i in np.round((subsets)*100).astype(int)])
plt.savefig("../figures/QC_recall-nvars.pdf", bbox_inches='tight', format='pdf', dpi=300)
plt.show()
```

```python
subsets = np.array([0, 10, 50, 100, 500, 1000, 1500, 2000, 2500, 3000])

C_std_all, C_mean_all = [], []
C_std_std_all, C_mean_std_all = [], []
runtime_all = []
shape_all = []
for patient in all_patients:
    print(patient)
    prefix = "AML_MutaSeq" if patient[0]=="P" else "AML_10X" if patient[0]=="A" else "LUNG"
    C_std, C_mean = [], []
    C_std_std, C_mean_std = [], []
    runtime = []
    for n_cycles in subsets:
        C_all = []
        runti = []
        try:
            adata = anndata.read_h5ad("../processed_data/robustness/"+patient +"_"+ str(n_cycles)+".h5ad")

            C_all = adata.uns["all_C"]

            std, mn = np.std(C_all, axis=0), np.mean(C_all, axis=0)

            C_std.append(np.mean(std)), C_mean.append(np.mean(mn))
            runtime.append(adata.uns["run_time"])
        except:
            C_std.append(np.nan), C_mean.append(np.nan)
            runtime.append(np.nan)

    C_std_all.append(C_std), C_mean_all.append(C_mean)
    runtime_all.append(runtime)
    shape_all.append(adata.shape[0]*adata.shape[1])

C_std_all, C_mean_all = np.array(C_std_all), np.array(C_mean_all)
C_std_std_all, C_mean_std_all = np.array(C_std_std_all), np.array(C_mean_std_all)
runtime_all = np.array(runtime_all)
```

```python
fig, axs = plt.subplots(3,3,figsize=(15,15), sharex=True)

for i, patient in enumerate(all_patients):
    c="pink" if patient[0]=="P" else "green" if patient[0]=="A" else "gold"
    yaxs =0 if patient[0]=="P" else 1 if patient[0]=="A" else 2
    
    for j, ax in enumerate(axs[yaxs]):
        x = subsets
        if j==0:
            y = (C_std_all[i]**2) / C_mean_all[i] 
            ax.set_ylabel("variance - mean ratio")
        elif j==1:
            y = C_mean_all[i] 
            ax.set_ylabel("mean")
        else:
            y=C_std_all[i]**2
            ax.set_ylabel("variance")
        ax.annotate(patient, (x[-1], y[-1]))  
        ax.scatter(x, y, c=c)
        ax.plot(x, y, c=c, alpha=.5)
        if patient[0]=="p":
            ax.set_xticks(subsets, subsets) 
            ax.set_xlabel("number of iterations")
        if patient[0]=="P":
            ax.set_title(("Variance mean ratio of " if j==0 else "Mean values in " if j==1 else "Variance of values in ")+"C")
        ax.set_yticks([.3,0.5, .7],[0.3,0.5, 0.7]) if j==1 else ax.set_yticks([0,.1,.2],[0,.1,.2])


plt.tight_layout()
plt.savefig("../figures/QC_mean_var_ratio.pdf", bbox_inches='tight', format='pdf', dpi=300)
plt.show()
```
